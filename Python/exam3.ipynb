```python
import numpy as np

a= np.array([1,2,3],int) # type ndarray 

np.linspace(0,2,10) # The numpy.linspace() function returns number spaces evenly w.r.t interval. 
#Similar to numpy.arange() function but instead of step it uses sample number.

np.logspace(1,10) # The numpy.logspace() function returns number spaces evenly w.r.t interval on a log scale. 

#arange 
print(np.arange(100, 1, -5)) #Return evenly spaced values within a given interval.
#When using a non-integer step, such as 0.1, it is often better to use numpy.linspace.
x2 = [(n ** 2) % 10 for n in np.arange(8)]
print(x2)

# arrays
a1 = np.array([1,2,-3,4,5])
a2 = np.array([10, 20, 30, 40, 50])
print (a1+a2)
print(a1-2)
print ("sine",np.sin(a1)) # Trigonometric sine, element-wise. 
print("Mean",np.mean (a1)) # mean of the array
print("Abs",np.abs(a1)) # negative to positive
print ("Sum",np.sum(a1))

# Logical comparison
a1 = np.array([1,2,-3,30,5])
a2 = np.array([10,20,30,20,4])
print(a1>a2)
print(a1<a2)
print(a1>=a2)
print(np.all(a1<a2)) # a2 all elements are greater than a1 return true
print(np.any(a1>a2)) # a1 any elements are greater than a2 return true

#coping array
a1 = np.arange(1,10,2)
print(a1)
a2 = a1.view() # copy of array 
print(a2)

# working over array iterating over array 
a1= np.arange(1,10)
print(a1)
print("A1 2 to 5",a1[2:5])
print("A1 0 to 8 two step",a1[0:8:2])
print ("A1 4 to last",a1[4:])

# ------------------------------
#multidimensional array
a1 = np.array([10,20,30])
a2 = np.array([[2,5,6],
[20, 30, 40]])
print(a2)
```
